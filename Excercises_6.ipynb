```python
from elasticsearch import Elasticsearch
import pickle

es = Elasticsearch()
```

```python
cats = pickle.load(open('code/Fountain categories web view/data/news_cats.pickle', 'rb'))
news = pickle.load(open('code/Fountain categories web view/data/texts_fontanka.pickle', 'rb'))
```

```python
es.indices.create(index='index', ignore=400)
```

```python
for i, n in enumerate(news):
    es.index(index='index',  doc_type='news', id=i, body={'title': cats[n[0]], 'content' : n[1]})
```

```python
cur = es.search(index='index', doc_type='news', q='госдума')
cur
```
